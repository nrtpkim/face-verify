### Prepare path

```python
'''
If you want to run .ipynb on vscode please install package below
pip install ipykernel
'''

import pandas as pd
import os
```

```python
# root_path = '1 with 1/1 with 1'
root_path = '../assets/593-person'
folder_path_arr = os.listdir(root_path)
print(folder_path_arr)
```

```python
### Prepare main path
no_img = 0
sum_img = 0

name_train = []
img_train = []

for folder_path in folder_path_arr:
    no_img = len(os.listdir(root_path+'/'+folder_path))
    sum_img = sum_img + no_img
    
    
    img_path_arr = os.listdir(root_path+'/'+folder_path)
    for img_path in img_path_arr:
        name_train.append(folder_path)
        img_train.append(img_path)
        # print(folder_path, img_path)
    
df = pd.DataFrame()

df["name_train"] = name_train
df["img_train"] = img_train
df  
```

```python
### Function for generate positive pair & negative pair
def prepare_df_fin(df_test, df_train, method):
    df_fin = pd.DataFrame()
    df_fin["name_train"] = df_train["name_train"].tolist()
    df_fin["img_train"] = df_train["img_train"].tolist()

    df_fin["name_test"] = df_test["name_train"].tolist()
    df_fin["img_test"] = df_test["img_train"].tolist()
    df_fin["check"] = (df_fin["name_train"] == df_fin["name_test"])
    df_fin = df_fin[df_fin['check']== method]
    df_fin = df_fin.drop_duplicates()
    df_fin = df_fin.reset_index(drop = True)
    
    return df_fin


def shuffle_data(df, df_train,df_test,size=20):
    # df_train = pd.DataFrame()
    for i in range(size):
        df_shuf = df.sample(frac=1)
        df_train = pd.concat([df_train,df_shuf])

    # df_test = pd.DataFrame()
    for i in range(size):
        df_shuf = df.sample(frac=1)
        df_test = pd.concat([df_test,df_shuf])
        
    
    return df_train, df_test
```

### Prepare data negative pairs

```python
df_train = pd.DataFrame()
df_test = pd.DataFrame()
df_train, df_test = shuffle_data(df, df_train,df_test, size=100)
df_negative_pair = prepare_df_fin(df_test, df_train, method=False)

print(df_negative_pair['check'].value_counts())
df_negative_pair
```

### Prepare data positive pairs

```python
from tqdm import tqdm
df_fin = pd.DataFrame()

df_train = pd.DataFrame()
df_test = pd.DataFrame()

for name in tqdm(folder_path_arr):
    df_focus = df[df['name_train']==name]
    df_train, df_test = shuffle_data(df_focus, df_train, df_test, size=200)
    
df_positive_pair = prepare_df_fin(df_test, df_train,method=True)

print(df_positive_pair['check'].value_counts())
df_positive_pair
```

```python
df_positive_pair = df_positive_pair[df_positive_pair['img_train']!=df_positive_pair['img_test']]
df_positive_pair = df_positive_pair.reset_index(drop=True)
df_positive_pair
```

### Combine dataset

```python
df_fin = pd.concat([df_negative_pair, df_positive_pair])
df_fin = df_fin.reset_index(drop=True)
df_fin
```

```python
df_fin.to_csv("../assets/data_clean_593.csv")
```
